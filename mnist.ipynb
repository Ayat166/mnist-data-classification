```python
y_train = pd.DataFrame(y_train, columns=['Label'])
y_train = y_train[(y_train["Label"] == 0) | (y_train["Label"] == 1)]
print(len(y_train))
y_train.head()
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def logistic_regression(X, y, lr, epochs):
    m, n = X.shape
    w = np.zeros((n, 1))
    b = 0
    loss=[]
    for epoch in range(epochs):
        z = np.dot(X, w) + b
        y_pred = sigmoid(z)
        cost = -1/m * np.sum(y*np.log(y_pred) + (1-y)*np.log(1-y_pred))
        dw = 1/m * np.dot(X.T, y_pred - y)
        db = 1/m * np.sum(y_pred - y)
        w = w - lr * dw
        b = b - lr * db
        loss.append(cost)
    return w, b,loss
```

```python
k = 10
lr= 0.001
fold_size = len(X_train) // k
accuracies = []
accuracy = 0
losses=[]
for i in range(k):
      val_start = i*fold_size
      val_end = (i+1)*fold_size
      X_val = X_train[val_start:val_end]
      y_val = y_train[val_start:val_end]
      X_train1 = pd.concat([X_train[:val_start], X_train[val_end:]])
      y_train1 = pd.concat([y_train[:val_start], y_train[val_end:]])
      w, b ,cost= logistic_regression(X_train1.values, y_train1.values, lr, epochs=1000)
      y_pred = sigmoid(np.dot(X_val.values, w) + b)
      y_pred = np.round(y_pred)
      accuracy += np.mean(y_pred == y_val.values)
      losses.append(cost)
accuracies.append(accuracy / k)
print(f"logistic_regression accuracy : {accuracies[-1]}")
mean_loss = np.mean(losses, axis=0)
plt.plot(mean_loss, label="lr = 0.001")
plt.title("Training Loss vs. Iterations")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
def L1_regularization(X, y, lr, lam, epochs):
    m, n = X.shape
    w = np.zeros((n, 1))
    b = 0
    cost = []
    for epoch in range(epochs):
        z = np.dot(X, w) + b
        y_pred = sigmoid(z)
        error = -1/m * np.sum(y*np.log(y_pred) + (1-y)*np.log(1-y_pred)) + lam*np.sum(np.abs(w))
         #= np.mean(np.square(y_pred - y)) + Lambda * np.sum(np.abs(w))
        cost.append(error)
        dw = 1/m * np.dot(X.T, y_pred - y) + lam*np.sign(w)
        db = 1/m * np.sum(y_pred - y)
        w = w - lr * dw
        b = b - lr * db
    return w, b, cost
```

```python
k = 10
fold_size = len(X_train) // k
lr = 0.001
lams = [0.1, 0.5]
accuracies = []
losses = []
for lam in lams:
    accuracy = 0
    for i in range(k):
        val_start = i*fold_size
        val_end = (i+1)*fold_size
        X_val = X_train[val_start:val_end]
        y_val = y_train[val_start:val_end]
        X_train1 = pd.concat([X_train[:val_start], X_train[val_end:]])
        y_train1 = pd.concat([y_train[:val_start], y_train[val_end:]])
        w, b ,cost = L1_regularization(X_train1.values, y_train1.values, lr, lam, epochs=1000)
        y_pred = sigmoid(np.dot(X_val.values, w) + b)
        y_pred = np.round(y_pred)
        accuracy += np.mean(y_pred == y_val.values)
        losses.append(cost)                          #record all error(loss)
    mean_loss = np.mean(losses, axis=0)  #calc mean difference losses
    plt.plot(mean_loss, label=f"Lambda: {lam}")
    accuracies.append(accuracy / k)
    print(f"Lambda {lam}:  L1 accuracy= {accuracies[-1]}")
# Add labels to the plot
plt.title("Loss curves for different learning rates")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
def mini_batch(X, y, lr, batch_size, epochs):
    m, n = X.shape
    w = np.zeros((n, 1))
    b = 0
    loss=[]
    for epoch in range(epochs):
        for i in range(0, m, batch_size):
            X_batch = X[i:i+batch_size]
            y_batch = y[i:i+batch_size]
            z = np.dot(X_batch, w) + b
            y_pred = sigmoid(z)
            cost = -1/batch_size * np.sum(y_batch*np.log(y_pred) + (1-y_batch)*np.log(1-y_pred))
            dw = 1/batch_size * np.dot(X_batch.T, y_pred - y_batch)
            db = 1/batch_size * np.sum(y_pred - y_batch)
            w = w - lr * dw
            b = b - lr * db
        loss.append(cost)
    return w, b,loss

```

```python
k = 10
fold_size = len(X_train) // k
lr = 0.001
batchs = [32, 64, 128]
accuracies = []
losses = []

for batch in batchs:
    accuracy = 0
    batch_losses = []  # Initialize losses list for recording costs

    for i in range(k):
        val_start = i * fold_size
        val_end = (i + 1) * fold_size
        X_val = X_train[val_start:val_end]
        y_val = y_train[val_start:val_end]
        X_train1 = pd.concat([X_train[:val_start], X_train[val_end:]])
        y_train1 = pd.concat([y_train[:val_start], y_train[val_end:]])
        w, b, cost = mini_batch(X_train1.values, y_train1.values, lr, batch, epochs=1000)
        y_pred = sigmoid(np.dot(X_val.values, w) + b)
        y_pred = np.round(y_pred)
        accuracy += np.mean(y_pred == y_val.values)
        batch_losses.append(cost)

    losses.append(batch_losses)
    accuracies.append(accuracy / k)
    print(f"batch {batch}:  batch accuracy= {accuracies[-1]}")

# Plotting the loss curves for all batch sizes
for i, batch in enumerate(batchs):
    mean_loss = np.mean(losses[i], axis=0)
    plt.plot(mean_loss, label=f"batch: {batch}")

# Add labels to the plot
plt.title("Loss curves for different batch sizes")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
def RMSProp(X, y, lr, epochs, beta=0.9, epsilon=1e-8):
    n, d = X.shape
    w = np.zeros((d, 1))
    b = 0
    cache_dw = np.zeros((d, 1))
    cache_db = 0
    cost=[]
    for epoch in range(epochs):
        y_pred = sigmoid(np.dot(X, w) + b)
        dw = (1/n) * np.dot(X.T, y_pred - y)
        db = (1/n) * np.sum(y_pred - y)
        cost.append(np.mean(-y * np.log(y_pred) - (1 - y) * np.log(1 - y_pred)))
        # Update cache
        cache_dw = beta * cache_dw + (1 - beta) * np.power(dw, 2)
        cache_db = beta * cache_db + (1 - beta) * np.power(db, 2)

        # Update parameters
        w = w - (lr / np.sqrt(cache_dw + epsilon)) * dw
        b = b - (lr / np.sqrt(cache_db + epsilon)) * db

    return w, b ,cost
```

```python
k = 10
fold_size = len(X_train) // k
lr = 0.001
accuracies = []
losses = []
accuracy = 0
for i in range(k):
    val_start = i*fold_size
    val_end = (i+1)*fold_size
    X_val = X_train[val_start:val_end]
    y_val = y_train[val_start:val_end]
    X_train1 = pd.concat([X_train[:val_start], X_train[val_end:]])
    y_train1 = pd.concat([y_train[:val_start], y_train[val_end:]])
    w, b,cost = RMSProp(X_train1.values, y_train1.values, lr, epochs=1000, beta = 0.9, epsilon = 1e-8)
    y_pred = sigmoid(np.dot(X_val.values, w) + b)
    y_pred = np.round(y_pred)
    accuracy += np.mean(y_pred == y_val.values)
    losses.append(cost)
mean_loss = np.mean(losses, axis=0)
accuracies.append(accuracy / k)
print(f"RMS_Prop accuracy : {accuracies[-1]}")
plt.plot(mean_loss, label="lr = 0.001")
plt.title("Training Loss vs. Iterations")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
def Adam(X, y, lr, epochs, beta1=0.9, beta2=0.999, epsilon=1e-8):
    m, n = X.shape
    w = np.zeros((n, 1))
    b = 0
    v = np.zeros((n, 1))
    s = np.zeros((n, 1))
    cost=[]
    for epoch in range(epochs):
        z = np.dot(X, w) + b
        a = sigmoid(z)
        dw = (1 / m) * np.dot(X.T, (a - y))
        db = (1 / m) * np.sum(a - y)
        v = beta1 * v + (1 - beta1) * dw
        s = beta2 * s + (1 - beta2) * np.square(dw)
        loss = np.mean((a - y) ** 2)
        cost.append(loss)
        v_corrected = v / (1 - beta1**(epoch+1))
        s_corrected = s / (1 - beta2**(epoch+1))
        w = w - lr * v_corrected / (np.sqrt(s_corrected) + epsilon)
        b = b - lr * db
    return w, b,cost

```

```python
k = 10
fold_size = len(X_train) // k
lr = 0.001
accuracies = []
accuracy = 0
losses=[]
for i in range(k):
    val_start = i*fold_size
    val_end = (i+1)*fold_size
    X_val = X_train[val_start:val_end]
    y_val = y_train[val_start:val_end]
    X_train1 = pd.concat([X_train[:val_start], X_train[val_end:]])
    y_train1 = pd.concat([y_train[:val_start], y_train[val_end:]])
    w, b ,cost= Adam(X_train1.values, y_train1.values.reshape(-1, 1), lr, epochs=1000, beta1=0.9, beta2=0.999, epsilon=1e-8)
    y_pred = sigmoid(np.dot(X_val.values, w) + b)
    y_pred = np.round(y_pred)
    accuracy += np.mean(y_pred == y_val.values)
    losses.append(cost)
mean_loss = np.mean(losses, axis=0)
accuracies.append(accuracy / k)
print(f"Adam accuracy : {accuracies[-1]}")
plt.plot(mean_loss, label="lr = 0.001")
plt.title("Training Loss vs. Iterations")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

*  Logistic Regression: as we working on binary classification data (0,1) as It
uses a logistic function to model the probability of a binary outcome.

*L1 Regularization: is a technique used to reduce the complexity of a model by adding a penalty term to the loss function( lam*np.sum(np.abs(w))). The penalty term is proportional to the absolute value of the model parameters.



* Mini Batch: Mini-batch is a technique used in optimization algorithms that involves splitting the training data into small batches and updating the model parameters based on each batch. This allows for faster convergence than using batch Logistic Regression , as it is less computationally expensive to calculate the gradient of a small batch than the entire dataset. Mini-batch is commonly used where large datasets are common, and efficient optimization is crucial for training deep neural networks.


* RMSProp: it uses a moving average of the squared gradients instead of the sum of the squared gradients to adjust the learning rate. This helps to reduce the impact of noisy or sparse gradients. RMSProp is also known to work well in non-convex optimization problems.



* Adam: It combines the benefits of both Momentum and RMSProp to provide adaptive learning rates that are specific to each parameter. It also uses momentum to accelerate convergence and avoid getting stuck in local minima. Adam is a popular choice for training deep neural networks due to its efficiency and robustness.
